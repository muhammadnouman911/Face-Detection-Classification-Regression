##Face Detection in image (MTCNN)

---



```python
!pip install mtcnn opencv-python

from mtcnn import MTCNN
import cv2
from google.colab.patches import cv2_imshow

# Load image
image = cv2.imread('/content/drive/MyDrive/dataset/faces/-wuxbgMRIWs_30.03_36.745044/frame_0010.jpg')  # Replace with your image path
image_rgb = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)

# Initialize detector
detector = MTCNN()

# Detect faces
faces = detector.detect_faces(image_rgb)

# Draw bounding boxes
for face in faces:
    x, y, w, h = face['box']
    cv2.rectangle(image, (x, y), (x+w, y+h), (0, 255, 0), 2)

cv2_imshow(image)
# cv2.waitKey(0) # waitKey and destroyAllWindows are not needed with cv2_imshow
# cv2.destroyAllWindows()
```

##Face Detection in Video (MTCNN)

---



```python
!pip install mtcnn opencv-python imageio

import cv2
from mtcnn import MTCNN
import imageio
from IPython.display import Video

# Load face detector
detector = MTCNN()

video_path = "/content/drive/MyDrive/dataset/video/GroLpQWzcPQ_235.0_238.666667.mp4"
cap = cv2.VideoCapture(video_path)

# Temporary output path
output_path = "/content/output_preview.mp4"
fps = int(cap.get(cv2.CAP_PROP_FPS))
width = int(cap.get(cv2.CAP_PROP_FRAME_WIDTH))
height = int(cap.get(cv2.CAP_PROP_FRAME_HEIGHT))

writer = imageio.get_writer(output_path, fps=fps)

while cap.isOpened():
    ret, frame = cap.read()
    if not ret:
        break

    rgb_frame = cv2.cvtColor(frame, cv2.COLOR_BGR2RGB)
    faces = detector.detect_faces(rgb_frame)

    # Draw bounding boxes
    for face in faces:
        x, y, w, h = face['box']
        cv2.rectangle(rgb_frame, (x, y), (x+w, y+h), (0, 255, 0), 2)
        cv2.putText(rgb_frame, "Face", (x, y-10),
                    cv2.FONT_HERSHEY_SIMPLEX, 0.9, (0, 255, 0), 2)

    writer.append_data(rgb_frame)  # Append frame to preview video

cap.release()
writer.close()

# Display the video directly
Video(output_path, embed=True)
```
